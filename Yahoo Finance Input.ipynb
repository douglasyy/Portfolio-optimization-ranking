```python
def get_key_stats(website):
#     url = 'https://finance.yahoo.com/quote/WDC/key-statistics?p=WDC'
    r = requests.get(website,headers ={'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'})

    result_df = pd.read_html(r.text)[0].rename(columns = {'Unnamed: 0': 'Title', 'As of Date: 9/27/2022Current' : 'Current'})
    return result_df
    
```

```python
r = requests.get('https://finance.yahoo.com/quote/ACN?p=ACN&.tsrc=fin-srch') 
soup = bs(r.content, 'lxml') 
# print(soup.prettify()[:300]) 
div_tag = soup.find('div', attrs={'id':'quote-summary'}) # all the information

# Left section
left = div_tag.find('div', {'data-test': 'left-summary-table'})

# <td class="Ta(end) Fw(600) Lh(14px)" data-test="FIFTY_TWO_WK_RANGE-value">256.20 - 417.37</td>
stats_1 = left.find_all('td', {'class': 'Ta(end) Fw(600) Lh(14px)'})

dict_0 = dict()
for element in stats_1:
    dict_0[element.get('data-test')[:-6]] = element.get_text()
    
# right section
right = div_tag.find('div', {'data-test': 'right-summary-table'})
# print(right)
stats_2 = right.find_all('td', {'class': 'Ta(end) Fw(600) Lh(14px)'})

for element in stats_2:
    dict_0[element.get('data-test')[:-6]] = element.get_text()
        
df_0 = pd.DataFrame.from_dict(dict_0, orient ='index') 
df_0 = df_0.rename(columns = {0: 'AAPL'})
df_0.index.name = 'id'
```
